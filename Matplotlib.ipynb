## 범례 표시하기

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0,3, .02)
c = np.exp(a)
d = c [::-1]

plt.plot(a,c,'r--', label = 'Model length')
plt.plot(a,d,'b:', label ='Data length')
plt.plot(a,c+d,'g', label ='Total message length')

plt.legend(loc = 'center right', shadow=True)
plt.show()
```

## 서브 플롯 그리기

### - subplot() 함수
- subplot()함수를 이용해서 하나의 그림에 여러개의 그래프를 그릴 수 있다.
- nrows, ncols, index의 인수를 이용하여 서브플롯의 위치를 설정한다.

```python
x1 = np.linspace(0.,5.,50)
x2 = np.linspace(0.,2.,50)

y1 = np.cos(2* np.pi *x1) + np.exp(-x1)
y2 = np.cos(2* np.pi *x2)

plt. subplot(2,1,1) # nrow =2, ncol=1, index=1
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time(s)')
plt.ylabel('Damped')

plt.subplot(2,1,2)
plt.plot(x2, y2, 'o-')
plt.title('2nd Graph')
plt.xlabel('time(s)')
plt.ylabel('undamped')

plt.tight_layout() # 서브플룻의 레이아웃을 자동으로 맞춰준다
plt.show()
```

### subplots 함수
fig 완전 바탕
axes 그래프 네모안
axes [0,0] 1사분면 방향 그래프

- subplots()함수는 Figure, Axes 객체를 반환한다.
- Axes 클래스 : 공간이 확보된 이미지 영역 정보를 가지고 있다.
    - 주요인자
   >- nrows, ncols : 서브플록의 행과 열 크기 지정
   >- sharex, sharey : boolean값(default : False), x/y축의 값 공유 여부

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1,4, sharey = True)
fig.set_size_inches((16,4))
axes[0].scatter(np.random.random(100), np.random.random(100), c='cyan', edgecolor='red', alpha =0.5)
axes[1].scatter(np.random.random(100), np.random.random(100), c='magenta', edgecolor='green', alpha =0.5)
axes[2].scatter(np.random.random(100), np.random.random(100), c='yellow', edgecolor='blue', alpha =0.5)
axes[3].scatter(np.random.random(100), np.random.random(100), c='white', edgecolor='black', alpha =0.5)
axes[0].set_title('subplot-0')
axes[1].set_title('subplot-1')
axes[2].set_title('subplot-2')
axes[3].set_title('subplot-3')

plt.show(0)
```

# 원 그래프 그리기

```python
data = [0.3, 0.35, 0.4]
label = ['python', 'Java', 'C']
colors = ['coral','dodgerblue','yellowgreen']

plt.pie(
    data, #조각의 크기를 결정하는 값
    labels=label, # 조각의 이름 표시
    explode=(0.1,0,0), # 조각을 중심으로부터 얼마나 떨어지게 할 것인지를 설정한다 (0.1 : 반지름의 10%만틈 벗어남)
    colors = colors, # 조각의 색깔
    autopct = '%.2r%%', #비율을 표시하는 형식 문자열
    shadow=True, # 그림자 표시 여부
    startangle=90, # 조각의 시작 부분을 각도로 표시, 해당 각도에서부터 반시계 방향으로 회전하면서 조각을 표시-기본값은 0, 3시방향
    counterclock=False, #반시계방향으로 순서대로 표시할지 결정 (기본값:True)
    rotatelabels=False #조각의 이름을 회전한다.(기본값:false)
)

plt.title('Popular language Ranking')
plt.show()
```

# 박스플롯 그리기
- 박스플롯은 데이터 집합의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(OUTLIER)가 있는지 검사하기 수월하다
- 최소값 : 제 1사분위에서 1.5IQR(inter quartile range : 사분위수범위, Q3-Q1)을 뺀 위치
- 제 1사분위(Q1): 중앙값 기준으로 하위 50%중의 중앙값, 전체 데이터 중 하위 25%에 해당
- 제 2사분위(Q2): 50%의 위치에 중앙값(MEDIAN)을 의미
- 제 3사분위(Q3): 중앙값 기준으로 상위 50%중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값: 제 3사분위에서 1.5IQR을 더한 위치
- 최대값 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0,2.0,1000) #평균 0 표편 2 1000
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)

plt.boxplot([data_a,data_b,data_c])
plt.ylim(-10.0,10.0)
plt.xlabel('Value')
plt.show()
```

```python
data_a = np.random.normal(0,2.0,1000)
print(f'min:{np.min(data_a):.2f},median:{np.median(data_a):2f},max:{np.max(data_a):2f}')
quantile = np.quantile(data_a,[0,0.25,0.5,0.75,1])
print(quantile)
IQR = quantile[3] - quantile [1]
print('IQR:',IQR)
print('Min:',quantile[1]-(1.5*IQR))
print('Max:',quantile[3]+(1.5*IQR))
```

```python
# 이상치 데이터 확인
np.random.seed(0)
spread = np.random.rand(50)*100  #[0,100) 데이터 50개
center = np.ones(25) * 50 #50 (중앙값 25개)
flier_high = np.random.rand(10) * 100 +100 # [100,200) 100이상 200미만 데이터 10개
flier_low = np.random.rand(10) * -100 # (-99, 0] 데이터 10개 
data = np.concatenate([spread, center, flier_high, flier_low])
print(data.shape)

plt.boxplot(data,0,'rx')
plt.grid(True)
plt.title('Box Plot')
plt.show()
```

```python
[실습] 매체별 광고 투자금액에 따른 상품 판매 효과 비교
- 선 그래프로 표현한다.
- 광고 매체는 tv와 google_ads 두 가지로 한다.
- x 값(투자금액)은 5~45사이의 5간격으로 만들어진 연속된 값 (tv와 google_ads 공동사용)
- tv의 y값은 (상품 판매 매출액) 1~64사이의 랜덤 정수 (오름차순 정렬)
- google_ad의 y값은 1~74사이의 랜덤 정수 (오름차순 정렬)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random  # random 모듈 임포트 추가

random.seed(10)
x_invest = list(range(5, 46, 5))
tv_y = [random.randint(1, 64) for i in range(9)]
google_y = [random.randint(1, 74) for i in range(9)]
tv_y.sort()
google_y.sort()

plt.plot(x_invest, tv_y, 'bs--', label='TV')
plt.plot(x_invest, google_y, 'ro--', label='GOOGLE ADS')

plt.title('광고 매체별 상품 판매 효과 비교')
plt.xlabel('광고 투자 금액', labelpad=10)
plt.ylabel('매출', labelpad=10)
plt.legend()
plt.show()
```
